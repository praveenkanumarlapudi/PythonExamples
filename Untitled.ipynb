```python
import os
os.system('kinit -kt  /users/s061634/s061634.keytab S061634@AETH.AETNA.COM')
from pyspark import SparkContext,SparkConf
from pyspark.sql import HiveContext
from pyspark.sql.types import Row
from datetime import date, timedelta
from pyspark.sql.types import *

def myFunc(s):
    oldColumns = s.schema.names
    newColumns = ["dst_ip","dst_port","src_ip","src_port","orig_ip_bytes","resp_ip_bytes","log_year"\
      ,"log_month","log_day"]
    s = reduce(lambda bro_src_ip_cnt_flat_df, idx: bro_src_ip_cnt_flat_df.withColumnRenamed(oldColumns[idx], newColumns[idx]), \
                           xrange(len(oldColumns)), s)
    s.printSchema() 
    try:
      r = s.select('dst_ip','dst_port','src_ip','src_port',int('orig_ip_bytes'),int('resp_ip_bytes'),'log_year'\
      ,'log_month','log_day')
      
    except ValueError, Attribute:
      r = s.select('dst_ip','dst_port','src_ip','src_port',int(0),int(0),'log_year','log_month','log_day')
    return r  
```

```python
queryString = 'SELECT dst_ip, dst_port, src_ip, src_port, orig_ip_bytes, resp_ip_bytes, log_year, log_month, log_day,\
                CONCAT(log_year,"-",log_month,"-",log_day) as log_date FROM shprd_sda_infra_enc.bro_conn_logs_raw\
                 WHERE uid is not NULL\
                 AND log_year  = "{0}"\
                 AND log_month = "{1}"\
                 AND log_day   = "{2}"'.format(processing_year,processing_month,processing_day)
```

```python
bro_conn_raw_df = hc.sql(queryString)
bro_conn_raw_rdd = bro_conn_raw_df.rdd
bro_conn_raw_rdd.cache()
```
